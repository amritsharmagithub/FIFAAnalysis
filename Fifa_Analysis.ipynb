## Distribution of Player skills

## Players with Left vs Right foot

```python
import matplotlib.pyplot as plt

#total count of left and right foot preferences
left_count = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right_count = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

# Calculate the total count of players
total_players = left_count + right_count

# Calculate the percentage of left and right foot preferences
left_percentage = (left_count / total_players) * 100
right_percentage = (right_count / total_players) * 100

#labels in Pie Chart
labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

# Plot the pie chart with percentages and labels inside
plt.pie([left_count, right_count], labels=labels, colors=colors, autopct='%1.1f%%', pctdistance=0.85)
plt.title('Foot Preference of FIFA players')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# smaller figure size
fig, axes = plt.subplots(1, 3, figsize=(8, 5))

new_df = fifa[fifa['Name'].str.contains('Cristiano Ronaldo|L. Messi', case=False)]


# Overall Performance distribution
sns.histplot(data=new_df, x='Overall', bins=20, kde=True, color='orange', ax=axes[0])
axes[0].set_title('Overall Performance Distribution')
axes[0].set_xlabel('Overall Rating')
axes[0].set_ylabel('Frequency')

# Plot Potential distribution
sns.histplot(data=new_df, x='Potential', bins=20, kde=True, color='green', ax=axes[1])
axes[1].set_title('Potential Distribution')
axes[1].set_xlabel('Potential Rating')
axes[1].set_ylabel('Frequency')

# Plot Composure distribution
sns.histplot(data=new_df, x='Composure', bins=20, kde=True, color='red', ax=axes[2])
axes[2].set_title('Composure Distribution')
axes[2].set_xlabel('Composure')
axes[2].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

## Comparison of player ratings in different clubs

```python
import matplotlib.pyplot as plt
import numpy as np

clubs = ['Real Madrid', 'Manchester United', 'FC Barcelona', 'Juventus', 'Paris Saint-Germain', 'Napoli']
club_data = []

for club in clubs:
    club_data.append(fifa.loc[fifa.Club == club]['Overall'])

plt.figure(figsize=(7, 3))

# Calculate the mean overall rating for each club
mean_ratings = [np.mean(data) for data in club_data]

# Plotting the horizontal bar chart
bars = plt.barh(clubs, mean_ratings, color='skyblue')
plt.xlabel('Average Overall Rating')
plt.title('Average Overall Player Ratings Among Football Clubs')
plt.grid(axis='x')

# Annotating each bar with its value
for bar, rating in zip(bars, mean_ratings):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{rating:.2f}', va='center')

plt.show()
```

## Different players Overall Performance

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Players to analyze
players_to_analyze = ['Cristiano Ronaldo', 'L. Messi', 'L. Suárez', 'Neymar Jr', 'E. Hazard']

# Filter the DataFrame for the specified players
players_data = fifa[fifa['Name'].isin(players_to_analyze)]

# Select numeric attributes for heatmap
attributes = ['Overall', 'Potential', 'Composure', 'Release Clause']

# Calculate the mean for each numeric attribute for each player
data = players_data.groupby('Name')[attributes].mean(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)
plt.title('Attributes Heatmap for different Players')
plt.xlabel('Attributes')
plt.ylabel('Players')
plt.show()
```
